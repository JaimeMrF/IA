```python
import pandas as pd


data = {
    "ID": [1, 2, 3, 4, 5],
    "Ciudad": ["Madrid", "Barcelona", "Sevilla", "Madrid", "Valencia"], 
    "Educación": ["Secundaria", "Primaria", "Universitaria", "Primaria", "Universitaria"],  
    "Género": ["F", "M", "F", "M", "F"],
}
df = pd.DataFrame(data)
print("Dataset original:")
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

ciudad_encoded = encoder.fit_transform(df[["Ciudad"]])

dfciudad = pd.DataFrame(ciudad_encoded, columns=encoder.get_feature_names_out(["Ciudad"]))
dfciudad.drop(columns="Ciudad_Valencia", axis=1, inplace=True)
dfc = df.join(dfciudad)

dfc.drop(columns="Ciudad", axis=1, inplace=True)
dfc
```

```python
dfsinciudades = df.drop(columns="Ciudad", axis=1)
dfsinciudades["Ciudad_Frecuencia"] = df["Ciudad"].map(df["Ciudad"].value_counts())
dfsinciudades
```
