```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, OrdinalEncoder


data = {
    "Nombre": ["Ana", "Luis", "Pedro", "María", "Sofía", "Carlos"],
    "Edad": [23, 35, None, 29, 40, 31],
    "Ciudad": ["Bogotá", "Medellín", "Bogotá", None, "Cali", "Cali"],
    "Ingreso_Mensual": [1500, 2500, 1800, None, 3000, 2200],
    "Compra_Frecuente": ["Sí", "No", "Sí", "No", None, "Sí"]
}

df = pd.DataFrame(data)

df
```

```python
df.Edad.fillna(df["Edad"].mean(), inplace=True)
df
```

```python
df.drop(columns="Nombre", axis=1, inplace=True)
df
```

```python
df.Ingreso_Mensual.fillna(df.Ingreso_Mensual.mean(), inplace=True)
df.Compra_Frecuente.bfill(inplace=True)
df

```

```python
df.Ciudad.ffill(inplace=True)
```

```python
encoder = OneHotEncoder(sparse_output=False)



ciudad_enconded = encoder.fit_transform(df[["Ciudad"]])
ciudad_df = pd.DataFrame(ciudad_enconded,columns=encoder.get_feature_names_out(["Ciudad"]))

dfc = df.join(ciudad_df)
dfc.drop(columns=['Ciudad', 'Ciudad_Medellín'], axis=1, inplace=True)
```

```python
dfc
```

```python
EncoderOrdinal = OrdinalEncoder()

compraEncoded = EncoderOrdinal.fit_transform(df[["Compra_Frecuente"]])

dfcCompra = pd.DataFrame(compraEncoded, columns=EncoderOrdinal.get_feature_names_out(['Compra_Frecuente']))

dfc["Compra_Frecuente"] = dfcCompra

dfc
```

```python
minmax = MinMaxScaler()

minmax = minmax.fit_transform(dfc[["Edad", "Ingreso_Mensual"]])


```

```python
newdf = pd.DataFrame(minmax, columns=['Edad', 'Ingreso_Mensual'])

dfc.join(newdf)

dfc
```
